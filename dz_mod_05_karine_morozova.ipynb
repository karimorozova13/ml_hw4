```python
import pickle

with open('./mod_05_topic_10_various_data.pkl', 'rb') as fl:
    data = pickle.load(fl)
    
autos = data['autos']
```

```python
autos['stroke_ratio'] = autos['stroke'] / autos['bore']
autos['make_and_style'] = autos['make'] + '_' + autos['body_style']
```

```python
X = autos.copy()
y = X.pop('price')
```

```python
cat_features = X.select_dtypes(include='object').columns

for col in cat_features:
    X[col], __doc__ = X[col].factorize()
```

```python
from sklearn.feature_selection import mutual_info_regression
import pandas as pd

mi = mutual_info_regression(X, y, 
                            discrete_features=X.columns.isin(
                                cat_features.to_list() +
                                ['num_of_doors',
                                 'num_of_cylinders']),
                            random_state=42)
    
mi = pd.Series(mi, name='MI scores', index=X.columns).sort_values()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.33, 
                                                    random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(learning_rate=0.3,
                                  subsample=0.75,
                                  max_features='sqrt',
                                  random_state=42).fit(X_train, y_train)
```

```python
feature_importances = pd.Series(model.feature_importances_, index=X.columns).sort_values()

comparison = pd.DataFrame({'MI scores': mi, 'Feature Importances': feature_importances})

# unification
comparison['MI scores rank'] = comparison['MI scores'].rank(pct=True)
comparison['Feature Importances rank'] = comparison['Feature Importances'].rank(pct=True)

# sort by MI
comparison = comparison.sort_values(by='MI scores', ascending=False)

print(comparison)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

comparison_long = pd.melt(comparison.reset_index(), id_vars='index', value_vars=['MI scores rank', 'Feature Importances rank'], 
                          var_name='Metric', value_name='Value')

sns.set_theme(style="whitegrid")

g = sns.catplot(
    data=comparison_long, kind="bar", orient='y',
    x="Value", y="index", hue="Metric",
    palette="dark", alpha=.6, height=6,
    aspect=2
)
g.despine(left=True)
g.set_xticklabels(rotation=90)
g.set_axis_labels("Features", "Scores")
g.legend.set_title("Metrics")
plt.title('Comparison of MI Scores and Feature Importances')
plt.show()
```

### Висновки

1. **MI Scores and Feature Importances:**

- Curb weight, engine size, та horsepower є найбільш важливими ознаками як за MI, так і за показником важливості. Це вказує на те, що ці ознаки мають значний вплив на цільову змінну як з точки зору MI, так і з точки зору Feature Importances.
  
- Highway_mpz та city_mpg показують високу взаємну інформацію, але їх важливість у моделі значно нижча. Це може вказувати на те, що ці ознаки мають нелінійний або складний взаємозв'язок з цільовою змінною, який модель не може легко виявити.
  
- Fuel system та make_and_style мають середні значення як за MI, так і за Feature Importances, що вказує на їхню певну значущість в обох випадках.

2. **Ознаки з високим MI, але низькою Feature Importances в моделі:**
   
- Highway_mpg має високу MI (0.870499), але низьку Feature Importances у моделі (0.24). Це може свідчити про те, що ця ознака добре пояснює цільову змінну, але модель не може повністю використати її потенціал.
  
- City_mpg також має високу MI (0.762674) та середню Feature Importances у моделі (0.56), що вказує на аналогічну ситуацію.

3. **Ознаки з низьким MI, але високою Feature Importances:**
   
- Drive_wheels має низьку MI (0.332490), але високу Feature Importances (0.84). Це вказує на те, що ця ознака є важливою для прогнозування цільової змінної, навіть якщо її взаємозв'язок з цільовою змінною не є очевидним.

4. **Інші:**
   
- Fuel type та engine location мають низькі значення як за MI, так і за Feature Importances, що свідчить про їхню низьку корисність для прогнозування цільової змінної і можуть не використовуватися для прогнозування.
  
- Num_of_doors має дуже низькі значення в обох метриках, що вказує на її незначний вплив на цільову змінну.

5. **Загальні висновки:**
   
- Високі значення MI не завжди корелюють з високою Feature Importances у моделі. Це вказує на те, що MI є корисним інструментом для початкової оцінки корисності ознак, але не завжди може передбачити їхню важливість у складних моделях.
  
- Feature Importances у моделі може бути кращим показником їхньої корисності для конкретної задачі прогнозування, оскільки вона враховує як взаємозв'язок з цільовою змінною, так і взаємодію з іншими ознаками.

Таким чином, комбіноване використання показника MI та Feature Importances у моделі дозволяє більш повно оцінити корисність кожної ознаки та зробити більш обґрунтовані рішення щодо їх включення в прогнозну модель.
